```python
# 958.Check Completeness of a Binary Tree 二叉树的完全性检验
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 临界节点：只有左子树或没有子树
class Solution(object):
    def isCompleteTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        queue =[root]
        flag = True
        while queue:
            node = queue.pop(0)
            if flag is False:
                if node.left or node.right:
                    return False
            else:
                # 左右都无，则临界
                if not node.left and not node.right:
                    flag = False
                # 左右都有
                elif node.left and node.right:
                    queue.append(node.left)
                    queue.append(node.right)
                # 有一个，是左，也临界
                elif node.left:
                    queue.append(node.left)
                    flag = False
                else:
                    return False
        return True        
```

```python
# 111.Minimum Depth of Binary Tree 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root:
            return 0
        queue = [root]
        depth = 1
        while queue:
            size = len(queue)
            for _ in range(size):
                node = queue.pop(0)
                if not node.left and not node.right:
                    return depth
                if node.left:
                    queue.appand(node.left)
                if node.right:
                    queue.appand(node.right)
            depth += 1
        return depth

```

```python
# 897. Increasing Order Search Tree 递增顺序搜索树
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def increasingBST(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        return self.traverse(root)

    def traverse(self, root):
        if not root:
            return None
        left = self.traverse(root.left)
        right = self.traverse(root.right)
        root.right = right
        if not left:
            return root

        tail = left
        while tail and tail.right:
            tail = tail.right
        tail.right = root
        root.left = None
        
        return left
```

```python
# 嵌套列表迭代器2
# """
# This is the interface that allows for creating nested lists.
# You should not implement it, or speculate about its implementation
# """
# 惰性迭代
class NestedInteger(object):
   def isInteger(self):
       """
       @return True if this NestedInteger holds a single integer, rather than a nested list.
       :rtype bool
       """

   def getInteger(self):
       """
       @return the single integer that this NestedInteger holds, if it holds a single integer
       Return None if this NestedInteger holds a nested list
       :rtype int
       """

   def getList(self):
       """
       @return the nested list that this NestedInteger holds, if it holds a nested list
       Return None if this NestedInteger holds a single integer
       :rtype List[NestedInteger]
       """

class NestedIterator(object):

    def __init__(self, nestedList):
        """
        Initialize your data structure here.
        :type nestedList: List[NestedInteger]
        """
        self.stack = []
        for i in range(len(nestedList)-1,-1,-1):
            self.stack.append(nestedList[i])

    def next(self):
        """
        :rtype: int
        """
        cur = self.stack.pop()
        return cur.getInteger()

    def hasNext(self):
        """
        :rtype: bool
        """
        while self.stack:
            cur = self.stack[-1]
            if cur.isInteger():
                return True
            self.stack.pop()
            for i in range(len(cur.getList()-1,-1,-1)):
                self.stack.append(cur.getList()[i])
        return False

# Your NestedIterator object will be instantiated and called as such:
# i, v = NestedIterator(nestedList), []
# while i.hasNext(): v.append(i.next())
```

```python
# N-ary Tree Level Order Traversal N叉树层序遍历
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
"""

class Solution(object):
    def levelOrder(self, root):
        """
        :type root: Node
        :rtype: List[List[int]]
        """
        if not root:
            return None
        result = []
        queue = [root]
        while queue:
            size = len(queue)
            level = []
            for _ in range(size):
                node = queue.pop(0)
                level.append(node.val)
                for child in node.children:
                    queue.append(child)
            result.append(level)
        return result
```

```python
# 124. Binary Tree Maximum Path Sum ⼆叉树中的最⼤路径和
# Binary Tree Maximum Path Sum
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def __init__(self):
        self.max_sum=-1001

    def maxPathSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        def connect(node):
            if node is None:
                return 0
            lp=connect(node.left)
            rp=connect(node.right)
            # 后序遍历位置更新当前位置的最大路径和one_path
            # 一个用来记录全局最大路径和，一个用来当返回值
            one_path=max(lp+node.val,rp+node.val,0,node.val)
            self.max_sum=max(self.max_sum,node.val+lp+rp)
            return one_path
        connect(root)
        return self.max_sum
```

```python
# 1245. Tree Diameter 树的直径（最小生成树）！！！！！！
# The idea is:
# 1. get children depth from bottom up
# 2. calculate local diameter
# 3. contribute to global diameter result
# 没有环的图都可以当成多叉树来遍历，因为可以以任意一个节点作为根节点
class Solution(object):
    def treeDiameter(self, edges):
        """
        :type edges: List[List[int]]
        :rtype: int
        """
        # 先把这个图写到字典里
        if not edges:
            return 0
        self.maxDiameter = 0
        self.tree = {}
        self.visited = set()
        for edge in edges:
            first = edge[0]
            second = edge[1]
            if not self.tree.has_key(first):
                self.tree[first] = []
            if not self.tree.has_key(second):
                self.tree[second] = []
            self.tree[first].append(second)
            self.tree[second].append(first)
        self.maxDepth(edges[0][1])
        return self.maxDiameter

    def maxDepth(self, root):
        if root in self.visited:
            # 被visited过说明是之前访问过的节点，不能继续延伸
            return 0
        self.visited.add(root)
        # 因为不知道子节点有几个，所以只记录最长的两支
        firstMaxDepth = 0
        secondMaxDepth = 0
        for child in self.tree[root]:
            childDepth = self.maxDepth(child)
            if childDepth >= firstMaxDepth:
                secondMaxDepth = firstMaxDepth
                firstMaxDepth = childDepth
            elif childDepth> secondMaxDepth:
                secondMaxDepth = childDepth
        myDiameter = firstMaxDepth + secondMaxDepth
        self.maxDiameter = max(self.maxDiameter,myDiameter)
        return 1+firstMaxDepth
```

```python
# 998. Maximum Binary Tree II 最⼤⼆叉树 II
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def insertIntoMaxTree(self, root, val):
        """
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode
        """
        self.node = TreeNode(val)
        return self.insert(root)

    def insert(self,root):
        if not root:
            return self.node
        if root.val < self.node.val:
            self.node.left = root
            return self.node
        else:
            root.right = self.insert(root.right)
            return root
```

```python
# 1740. Find Distance in a Binary Tree 找到⼆叉树中的距离
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def findDistance(self, root, p, q):
        """
        :type root: TreeNode
        :type p: int
        :type q: int
        :rtype: int
        """
        self.p = p
        self.q = q
        self.path = 0
        self.LCA(root)
        return self.path
    # 向下递的时候不更新是否含pq
    # 向上归时才会携带当前节点所在枝是否含有pq的信息
    # found_p和found_q is local variable of each node, if the target node is found 
    # the information will be passed to the perent node from bottom up
    def LCA(self,root):
        found_p = False
        found_q = False
        if not root:
            return (False,False)
        
        left = self.LCA(root.left)
        right = self.LCA(root.right)
        
        found_p = left[0] or right[0]
        found_q = left[1] or right[1]
        if root.val ==self.p:
            found_p = True
        if root.val == self.q:
            found_q = True
        if found_p and found_q:
            return (False,False)
        if found_p or found_q:
            self.path +=1
        
        return (found_p,found_q)
```
