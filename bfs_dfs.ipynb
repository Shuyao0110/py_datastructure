```python
# 102.Binary Tree Level Order Traversal
# 给你⼀个⼆叉树，请你返回其按层序遍历得到的节点值。（即逐层地，从左到右访问所有节点）。
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root:
            return None
        results = [] 
        queue = [root]
        while queue:
            size = len(queue)
            level = []
            for _ in range(size):
                node = queue.pop(0)
                level.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            results.append(level)
        return results
```

```python
# 1161.Maximum Level Sum of a Binary Tree 最大层内元素和
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxLevelSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None:
            return None
        queue = [root]
        results = []
        while queue:
            level_sum = 0
            size = len(queue)
            for _ in range(size):
                node = queue.pop(0)
                level_sum += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            results.append(level_sum)
        return results.index(max(results))+1
```

```python
# 637.Average of Levels in Binary Tree 二叉树层平均值
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 用双层循环，从两个维度遍历一棵二叉树，层数和每层的长度
class Solution(object):
    def averageOfLevels(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        queue = [root]
        results = []
        while queue:
            size = len(queue)
            level = 0
            for _ in range(size):
                node = queue.pop(0)
                level += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            results.append(1.0*level/size)
        return results
```

```python
# 958.Check Completeness of a Binary Tree 二叉树的完全性检验
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 临界节点：只有左子树或没有子树
class Solution(object):
    def isCompleteTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        queue =[root]
        flag = True
        while queue:
            node = queue.pop(0)
            if flag is False:
                if node.left or node.right:
                    return False
            else:
                # 左右都无，则临界
                if not node.left and not node.right:
                    flag = False
                # 左右都有
                elif node.left and node.right:
                    queue.append(node.left)
                    queue.append(node.right)
                # 有一个，是左，也临界
                elif node.left:
                    queue.append(node.left)
                    flag = False
                else:
                    return False
        return True        
```

```python
# 111.Minimum Depth of Binary Tree 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root:
            return 0
        queue = [root]
        depth = 1
        while queue:
            size = len(queue)
            for _ in range(size):
                node = queue.pop(0)
                if not node.left and not node.right:
                    return depth
                if node.left:
                    queue.appand(node.left)
                if node.right:
                    queue.appand(node.right)
            depth += 1
        return depth

```

```python
# 114. Flatten Binary Tree to Linked List
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def flatten(self, root):
        """
        :type root: TreeNode
        :rtype: None Do not return anything, modify root in-place instead.
        """
        if not root:
            return
        # 先把左右子树拉平
        self.flatten(root.left)
        self.flatten(root.right)

        temp = root.right
        root.right = root.left
        root.left = None
        while root.right:
            root = root.right
        root.right = temp
        
```
