```python
# 94.Binary Tree Inorder Traversal
# 给定⼀个⼆叉树的根节点 root，返回它的 中序 遍历。
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
        
class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        self.result=[]
        self.traverse(root)
        return self.result
        
    def traverse(self, root):
        if root is None:
            return
        self.traverse(root.left)
        self.result.append(root.val)
        self.traverse(root.right)
        
```

```python
# 100.Same Tree 
# 给你两棵⼆叉树的根节点 p 和 q，编写⼀个函数来检验这两棵树是否相同。
# 如果两个树在结构上相同，并且节点具有相同的值，则认为它们是相同的。
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def __init__(self) :
        self.flag=True
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """
        if self.flag is False:
            return self.flag
        if p is None and q is None:
            return self.flag
        if p is None or q is None:
            self.flag=False
            return self.flag
        if p.val == q.val:
            self.isSameTree(p.left,q.left)
            self.isSameTree(p.right,q.right)
            return self.flag
        self.flag=False
        return self.flag
```

```python
# 572.Subtree of Another Tree
# 给你两棵⼆叉树 root 和 subRoot。
# 检验 root 中是否包含和 subRoot 具有相同结构和节点值的⼦树。
# 如果存在，返回 true；否则，返回 false。
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def __init__(self) :
        self.flag=True
    def isSubtree(self, root, subRoot):
        """
        :type root: TreeNode
        :type subRoot: TreeNode
        :rtype: bool
        """
        if root is None:
            return False
        self.subRoot = subRoot
        self.flag=False
        self.traverse(root)
        return self.flag
        
    def traverse(self, root):
        if self.flag is True:
            return
        if root is None:
            return
        self.flag=True
        self.isSameTree(root, self.subRoot)
        self.traverse(root.left)
        self.traverse(root.right)
        
    def isSameTree(self,p,q):
        if self.flag is False:
            return self.flag
        if p is None and q is None:
            return self.flag
        if p is None or q is None:
            self.flag=False
            return self.flag
        if p.val == q.val:
            self.isSameTree(p.left,q.left)
            self.isSameTree(p.right,q.right)
            return self.flag
        self.flag=False
        return self.flag
```

```python
# 102.Binary Tree Level Order Traversal
# 给你⼀个⼆叉树，请你返回其按层序遍历得到的节点值。（即逐层地，从左到右访问所有节点）。
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root:
            return None
        results = [] 
        queue = [root]
        while queue:
            size = len(queue)
            level = []
            for _ in range(size):
                node = queue.pop(0)
                level.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            results.append(level)
        return results
```

```python
# 1161.Maximum Level Sum of a Binary Tree 最大层内元素和
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxLevelSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None:
            return None
        queue = [root]
        results = []
        while queue:
            level_sum = 0
            size = len(queue)
            for _ in range(size):
                node = queue.pop(0)
                level_sum += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            results.append(level_sum)
        return results.index(max(results))+1
```

```python
# 637.Average of Levels in Binary Tree 二叉树层平均值
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 用双层循环，从两个维度遍历一棵二叉树，层数和每层的长度
class Solution(object):
    def averageOfLevels(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        queue = [root]
        results = []
        while queue:
            size = len(queue)
            level = 0
            for _ in range(size):
                node = queue.pop(0)
                level += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            results.append(1.0*level/size)
        return results
```

```python
# 919. Complete Binary Tree Inserter 完全二叉树插入器
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class CBTInserter(object):

    def __init__(self, root):
        """
        :type root: TreeNode
        """
        self.root = root
        self.queue = [root]
        while self.queue:
            node = self.queue.pop(0)
            if node.left:
                self.queue.append(node.left)
            if node.right:
                self.queue.append(node.right)
            if not node.left or not node.right:
                self.temp = node
                return

    def insert(self, val):
        """
        :type val: int
        :rtype: int
        """
        cur = TreeNode(val)
        self.queue.append(cur)
        if not self.temp.left:
            self.temp.left = cur
            return self.temp.val
        else:
            self.temp.right = cur
            parent = self.temp.val
            self.temp = self.queue.pop(0)
            return parent
        

    def get_root(self):
        """
        :rtype: TreeNode
        """
        return self.root
        
# Your CBTInserter object will be instantiated and called as such:
# obj = CBTInserter(root)
# param_1 = obj.insert(val)
# param_2 = obj.get_root()
```

```python
# 958.Check Completeness of a Binary Tree 二叉树的完全性检验
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 临界节点：只有左子树或没有子树
class Solution(object):
    def isCompleteTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        queue =[root]
        flag = True
        while queue:
            node = queue.pop(0)
            if flag is False:
                if node.left or node.right:
                    return False
            else:
                # 左右都无，则临界
                if not node.left and not node.right:
                    flag = False
                # 左右都有
                elif node.left and node.right:
                    queue.append(node.left)
                    queue.append(node.right)
                # 有一个，是左，也临界
                elif node.left:
                    queue.append(node.left)
                    flag = False
                else:
                    return False
        return True        
```

```python
# 104. Maximum Depth of Binary Tree
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.depth = 0
        self.res = 0
        self.traverse(root)
        return self.res
        
    def traverse(self,root):
        if root is None:
            return
        self.depth += 1
        self.res = max(self.depth,self.res)
        self.traverse(root.left)
        self.traverse(root.right)
        self.depth -=1
        
```

```python
# Time Needed to Inform All Employees通知所有员工所需时间
class Solution(object):
    def numOfMinutes(self, n, headID, manager, informTime):
        """
        :type n: int
        :type headID: int
        :type manager: List[int]
        :type informTime: List[int]
        :rtype: int
        """
        self.tree = {}
        self.informTime = informTime
        for i in range(n):
            parent = manager[i]
            if not self.tree[parent]:
                self.tree[parent] = []
            self.tree[parent].append(i)
            
    def maxDepth(self,root):
        if root is None:
            return
        subMaxDepth = 0
        for child in self.tree[root]:
            subMaxDepth = max(subMaxDepth,self.maxDepth(child))
        return self.informTime[root]+subMaxDepth
```
